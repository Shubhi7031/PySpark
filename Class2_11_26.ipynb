```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
```

```python
sc=SparkContext(master='local',appName='Revision')
```

```python
rdd=sc.textFile('test.txt')
rdd.collect()
```

```python
rdd1=rdd.map(lambda x: (x,1))
```

```python
rdd1.collect()
```

```python
rdd2=rdd1.reduceByKey(lambda x,y:x+y)
print(rdd2.collect())
```

```python
data=sc.textFile("sample2.json.json")
print(data.collect())
```

```python
data1=data.map(lambda x:(x,1))
print(data1.collect())
```

```python
data2=data1.reduceByKey(lambda x,y:x+y)
data2.collect()
```

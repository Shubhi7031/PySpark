##Change to RDD

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('RDD_revision').getOrCreate()
sc=spark.sparkContext
list1= [1,2,2,4,5,6,7,9]
print(type(list1))
rdd1= sc.parallelize(list1)
print(type(rdd1))
```

```python
##create new rdd as >=2 from rdd1
```

```python
rdd2= rdd1.filter(lambda x:x>2)
rdd2.collect()
```

```python
list2= [24,25,48,69,12,40,29]
rdd1=sc.parallelize(list2)
rdd2= rdd1.filter(lambda x: x<50)
rdd2.collect()
```

```python
##create a new by subtracting 10 from each value from rdd1
```

```python
rdd2=rdd1.map(lambda x:x-10)
rdd2.collect()
```

```python
data= ['pyspark session is going on','rdd concept started']
rdd1=sc.parallelize(data)

print("map output")
rdd2=rdd1.map(lambda x:x.split())
print(rdd2.collect())

print("flatmap output")
rdd3=rdd1.flatMap(lambda x:x.split())

rdd3.collect()
```

```python
##count(),first(),take()
```

```python
list1= [10,20,30,40,50,0,1,2,3,4,5,6,7,8]
rdd1=sc.parallelize(list1)

print(rdd1.count())
print(rdd1.first())
print(rdd1.take(2))
print(rdd1.take(5))
```

```python
rdd2=rdd1.sortBy(lambda x:x,ascending='True')
rdd2.collect()
```

```python
list1=[28,12,36,22,11,19,30]
rdd1=sc.parallelize(list1)

rdd2=rdd1.sortBy(lambda x:x, ascending= False)

rdd2.collect()
```

```python
rdd2=rdd1.fold(0,lambda x,y:x+y)
print(rdd2)
```
